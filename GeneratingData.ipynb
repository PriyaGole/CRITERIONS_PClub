```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

X1=10*np.random.rand(100,1)
y=15+4*X1+3*np.random.randn(100,1)
```

```python
plt.scatter(X1, y,s=10)

plt.xlabel('X')
plt.ylabel('y')

plt.title('Data')

plt.show()
```

```python
dfX=pd.DataFrame(X1)
dfX.rename(columns={0: 'X1'}, inplace=True)
dfX.to_csv("DataX.csv",index=False)
dfy=pd.DataFrame(y)
dfy.to_csv("DataY.csv",index=False)
```

```python
X1=np.random.uniform(-10,10,100)
y=1+1*X1+2*X1*X1+10*np.random.uniform(0,1,100)
```

```python
plt.scatter(X1, y,s=10)

plt.xlabel('X')
plt.ylabel('y')

plt.title('Data')

plt.show()
```

```python
dfX=pd.DataFrame(X1)
dfX.rename(columns={0: 'X1'}, inplace=True)
dfX.to_csv("DataX1.csv",index=False)
dfy=pd.DataFrame(y)
dfy.to_csv("DataY1.csv",index=False)
```
